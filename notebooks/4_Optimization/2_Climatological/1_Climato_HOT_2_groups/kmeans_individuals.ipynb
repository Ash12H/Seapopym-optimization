```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.cluster import KMeans
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
import plotly.express as px
from sklearn import preprocessing
```

```python
ranges = [
    [0, 0.5],
    [0, 50],
    [-1, 0],
    [0, 500],
    [-0.4, 0],
    [0, 1],
    [0, 50],
    [-1, 0],
    [0, 500],
    [-0.4, 0],
]
```

```python
data = pd.read_json("./SeapoPym_HOT_climato_obs_npp_opti_all_parameters_2_groups_logbook.json", orient="table")
data = data.sort_values(by="fitness", ascending=True)
data
```

```python
SELECTED = 5_000

data_post_prod = data.reset_index().query("previous_generation == False")
data_post_prod = data_post_prod.drop(columns=["generation", "previous_generation", "individual"])
# data_post_prod = data_post_prod.head(SELECTED)
data_post_prod
```

```python
px.parallel_coordinates(
    data_post_prod.head(2000),
    dimensions=data_post_prod.columns[:-2],
    color="fitness_final",
    color_continuous_scale=px.colors.diverging.Tealrose,
)
```

```python
weights = data_post_prod["fitness_final"]
weights = weights + np.abs(weights.min())
weights
```

```python
kmeans = KMeans(n_clusters=2)

kmeans.fit(data_post_prod.iloc[:, :-2].to_numpy(), sample_weight=weights)
```

```python
kmeans_2 = pd.DataFrame(kmeans.cluster_centers_, columns=data_post_prod.columns[:-2])
kmeans_2
```

```python
dimensions = []
for param_name, param_range in zip(kmeans_2.columns, ranges):
    dimensions.append(
        {
            "range": param_range,
            "label": param_name,
            "values": kmeans_2[param_name].to_list(),
        },
    )
```

```python
fig = go.Figure(data=go.Parcoords(dimensions=dimensions))
fig.show()
```
