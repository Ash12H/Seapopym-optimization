```python
import xarray as xr
import cf_xarray
import hvplot.xarray
```

```python
cafe_data = xr.open_dataset("../data/0_raw/CAFE_NPP_8D_25KM_1998_2022.nc")
cafe_data = cafe_data.rename({"lat": "latitude", "lon": "longitude"})
cafe_data["CAFE"] = cafe_data["CAFE"].assign_attrs({"units": "mg m-2 day-1", "weight": "carbon"})
cafe_data
```

```python
papa_obs = xr.open_dataset("../data/1_products/Papa_obs.zarr", engine="zarr")
calcofi_obs = xr.open_dataset("../data/1_products/Calcofi_obs.zarr", engine="zarr")
```

## HOT

---


```python
hot_obs = xr.open_dataset("../data/1_products/Hot_obs.zarr", engine="zarr")
hot_obs
```

```python
cafe_data_hot = cafe_data.interp_like(hot_obs[["latitude", "longitude"]])
cafe_data_hot
```

```python
cafe_data_hot.CAFE.hvplot.line(
    x="time",
    y="CAFE",
    groupby=["latitude", "longitude"],
    width=800,
    height=400,
    title="CAFE at HOT",
)
```

```python
cafe_data_hot.to_zarr("../data/1_products/Hot_cafe.zarr")
```
