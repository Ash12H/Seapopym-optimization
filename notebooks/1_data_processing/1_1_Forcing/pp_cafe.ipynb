```python
import xarray as xr
import cf_xarray
```

```python
cafe_data = xr.open_dataset("/Users/ash/Downloads/CAFE_NPP_8D_25KM_1998_2022.nc")
```

```python
cafe_data
```

## HOT

---


```python
LATITUDE_HOT = 22.75
LONGITUDE_HOT = -158
```

```python
cafe_data_hot = cafe_data.cf.sel(lat=[LATITUDE_HOT], lon=[LONGITUDE_HOT], method="nearest")
cafe_data_hot = cafe_data_hot.rename({"lat": "latitude", "lon": "longitude"})
cafe_data_hot = cafe_data_hot.assign_coords({"latitude": [LATITUDE_HOT], "longitude": [LONGITUDE_HOT]})
cafe_data_hot.latitude.attrs = {"units": "degrees_north", "long_name": "latitude", "axis": "Y"}
cafe_data_hot.longitude.attrs = {"units": "degrees_east", "long_name": "longitude", "axis": "X"}
cafe_data_hot["CAFE"] = cafe_data_hot["CAFE"].assign_attrs({"units": "mg m-2 day-1"})
cafe_data_hot
```

```python
cafe_data_hot.to_zarr("./Hot_cafe.zarr")
```
