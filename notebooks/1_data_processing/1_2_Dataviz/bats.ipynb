```python
import xarray as xr
import matplotlib.pyplot as plt
```

# CMEMS


```python
data = xr.load_dataset("../1_1_Forcing/products/Bats_cmems.zarr", engine="zarr")
data.T.isel(depth=0).plot()
```

# CAFE


```python
npp = xr.load_dataset("../1_1_Forcing/products/Bats_cafe.zarr", engine="zarr")
npp.CAFE.plot()
```

# OBSERVATIONS


```python
obs = xr.load_dataset("../1_1_Forcing/products/Bats_obs.zarr", engine="zarr")
obs.day.dropna("time").plot(label="day")
obs.night.dropna("time").plot(label="night")
plt.legend()
```
