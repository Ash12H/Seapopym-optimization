```python
import xarray as xr
import matplotlib.pyplot as plt
import plotly.express as px
```

# CMEMS


```python
data = xr.load_dataset("../1_1_Forcing/products/Hot_cmems.zarr", engine="zarr")
data.T.isel(depth=0).plot()
```

# CAFE


```python
npp = xr.load_dataset("../1_1_Forcing/products/Hot_cafe.zarr", engine="zarr")
npp.CAFE.plot()
```

# OBSERVATIONS


```python
obs = xr.load_dataset("../1_1_Forcing/products/Hot_obs.zarr", engine="zarr")
obs.day.dropna("time").plot(label="day")
obs.night.dropna("time").plot(label="night")
plt.legend()
```

# NPP CAFE vs VGPM


```python
fig = px.line()
fig.add_scatter(x=npp.time, y=npp.CAFE.isel(latitude=0, longitude=0), mode="lines", name="CAFE")
fig.add_scatter(x=data.time, y=data.npp.isel(latitude=0, longitude=0), mode="lines", name="VGPM")
fig.show()
```
